### Data Manipulation

```python
df.rename({'details':'configuration','title':'product_name'}, axis = 1, inplace = True)
```

```python
df
```

```python
df['website'] = 'Flipkart'
```

```python
df
```

```python
import numpy as np

def extract_rating(value):
    if value:
        value_0 = value.split(' ')[0]
        value_0 = value_0.replace(',','')
        return int(value_0)
    else:
        return np.nan

df['ratings'] = df['rr'].apply(extract_rating)
```

```python
df
```

```python

def extract_reviews(value):
    if value:
        value_0 = value.split()[-2]
        value_0 = value_0.replace(',','')
        return int(value_0)
    else:
        return np.nan

df['reviews'] = df['rr'].apply(extract_reviews)
```

```python
df
```

```python
df.drop(columns=['rr'], inplace = True)
```

```python
df
```

```python
# Task 1 : create a column brand and store the brand name in it taken from the first column
def extract_brand(value):
    if value:
        value_0 = value.split()[0]
        value_0 = value_0.replace(',','')
        return value_0
    else:
        return np.nan

df['brand'] = df['product_name'].apply(extract_brand)
```

```python
df
```

```python
df['price'] = df['price'].apply(lambda x: x.replace(',','').replace('₹',''))
```

```python
df
```

```python
df['link'] = df['link'].apply(lambda x: 'https://www.flipkart.com'+x)
```

```python
df
```

```python
df['price'] = df['price'].astype(int)
```
